## Notebook to examine a few topics timeseries (counts).


```python
import pandas as pd
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read in data file, each row is a topic number (index off by one)
which = '19'
timeseries_data = pd.read_csv(f"../data/topic_years_19_2_125_rec1.csv", index_col=False)
```

```python
# grab only columns we need
data = timeseries_data.drop('Unnamed: 0', axis=1)
#data
```

```python
def get_topic_data(data:pd.DataFrame, topic:int):

    topic_data = data.loc[topic]
    x = [ int(i) for i in topic_data.index]
    y = [ float(v) for v in topic_data.values]

    return [x, y]
```

```python
#plt.plot(df_topic)
def plotit(data:list, t:int):
    
    x = data[0]
    y = data [1]
    plt.plot(x, y, "k", label='Data')
    # plt.gca().invert_xaxis()
    #plt.plot(x_avg, y_avg, "ro", label='Avg Data')
    #plt.text(max_x +10, 0.9*max_y, f"Top avg %5.3f" % max_y_avg)
 
    plt.title(f"Topic {t} counts")
    plt.xlabel("Year")
    plt.ylabel(f"Counts")
    #plt.ylim(0,1.05)
    #plt.xlim(-1.,-5.1)
    plt.show()
```

```python
def plot3(data:dict):
    
    size = len(data.keys())
    
    #fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig, ax = plt.subplots(1, size)
    
    i = 0
    for k in data.keys():
          
        x = data[k][0]
        y = data[k][1]
        ax[i].plot(x, y)
        ax[i].set_xlabel("Year")
        i = i + 1

    #plt.ylim(0,150)
    ax[0].set_ylabel(f"TCS")
    
    #ax[0].text(2004, 150, '(a)', style='italic')
    #ax[0].set_ylim([40,160])
    #ax[1].text(2004, 150, '(b)', style='italic')
    #ax[1].set_ylim([40,160])
    #ax[2].text(2004, 150, '(c)', style='italic')
    #ax[2].set_ylim([40,160])
    
    # write to disk for distribution
    # fig = plt.gcf()
    fig.set_size_inches(14.0, 6.0)
    fig.subplots_adjust(left=0.15)
    
    plt.savefig('../sample_topic_timeseries.jpg', transparent=False, dpi=600)
```
